```python
import networkx as nx

# Load as UNDIRECTED graph (removes duplicate reciprocal edges)
G = nx.read_edgelist(
    "data/Email-Enron.txt",
    create_using=nx.Graph(),
    nodetype=int,
    comments="#"
)

print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

# Show first 10 edges
print("\nFirst 10 edges:")
for i, edge in enumerate(G.edges()):
    if i >= 10:
        break
    print(edge)
```

```python
print("\nGraph info:")
print("Graph type:", type(G))
print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
print("Is directed:", G.is_directed())
print("Density:", nx.density(G))


degrees = dict(G.degree())
print("Max degree:", max(degrees.values()))
print("Average degree:", sum(degrees.values()) / len(degrees))
```

```python
import matplotlib.pyplot as plt

# Take largest connected component
largest_cc = max(nx.connected_components(G), key=len)
G_sub = G.subgraph(list(largest_cc)[:200])  # only first 200 nodes

plt.figure(figsize=(8,8))
nx.draw(G_sub, node_size=10, with_labels=False)
plt.show()
```
